```python
import pandas as pd
import numpy as np
boston=pd.read_csv('BostonHousing.csv')
```

```python
boston.head(10)
```

```python
import statsmodels.api as sm
```

```python
target=pd.DataFrame(data=boston.values[:,13],columns=["MDEV"])
target
```

```python
x=boston['lstat']
y=target['MDEV']
x=sm.add_constant(x)
```

```python
model=sm.OLS(y,x).fit()
prediction=model.predict(x)
```

```python
model.summary()
```
